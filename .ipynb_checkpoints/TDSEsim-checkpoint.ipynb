```python
import numpy as np
import cmath as cm

dx = .01
dt = .01
n = 1000

psi0 = np.zeros(n)
psi1 = np.zeros(n)
u = np.zeros(n)
u[1] = 10000
u[n-2]= 10000
x = np.linspace(0,n*dx,n)

def timestep():
    for k in range(0,n):
        psi1[k]=psi0[k]
    
    for i in range(0,20):
        for k in range(1,n-1):
            denom = u[k] - 1j/dt - 2/(dx**2)
            numer = -(psi1[k+1]+psi1[k-1])/(dx**2) + 1j*psi0[k]/dt
            psi1[k] = numer/denom

```

```python
psi0 = np.sin(x) + 1j*np.cos(x)
```

```python
import matplotlib.pyplot as plt
plt.clf()
plt.plot(x,np.real(psi0),x,np.imag(psi0))
plt.show()
```

```python
from IPython.display import HTML
import matplotlib.animation as animation

plt.clf()
fig, ax = plt.subplots(1,1,figsize=(12,4))
real_part, = ax.plot(x,np.real(psi0))
imag_part, = ax.plot(x,np.imag(psi0))

def update(n):
    if(n>3):
        global psi0, psi1
        timestep()
        psi0, psi1 = psi1, psi0
        real_part.set_data(x,np.real(psi0))
        imag_part.set_data(x,np.imag(psi0))
        return fig
    
anim = animation.FuncAnimation(fig, update,frames=100, interval=100, blit=False)
video = anim.to_html5_video()
HTML(video)
```
